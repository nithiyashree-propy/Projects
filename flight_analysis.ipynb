```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df
# Replace 'Unnamed: 11' and 'Unnamed: 12' with actual data or drop if unnecessary
df = df.drop(columns=['Unnamed: 11', 'Unnamed: 12'], errors='ignore')

# 1. Top Airlines by Number of Flights
airline_counts = df['airline'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=airline_counts.index, y=airline_counts.values, palette='viridis')
plt.title('Top 10 Airlines by Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

```python
df['price'] = df['price'].str.replace(',', '').astype(float)
```

```python
avg_price_class = df.groupby('class')['price'].mean().sort_values()
plt.figure(figsize=(8, 6))
sns.barplot(x=avg_price_class.index, y=avg_price_class.values, palette='coolwarm')
plt.title('Average Flight Price by Class')
plt.show()
```

```python
# 3. Most Frequent Flight Routes
df['route'] = df['from'] + ' to ' + df['to']
route_counts = df['route'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=route_counts.index, y=route_counts.values, palette='Blues_r')
plt.title('Top 10 Most Frequent Flight Routes')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Flight Duration Analysis
plt.figure(figsize=(10, 6))
sns.histplot(df['duration'], kde=True, color='green')
plt.title('Flight Duration Distribution')
plt.xlabel('Duration (Minutes)')
plt.show()
```

```python
# 5. Peak Departure Times
df['dep_hour'] = pd.to_datetime(df['dep_time']).dt.hour
plt.figure(figsize=(10, 6))
sns.histplot(df['dep_hour'], bins=24, color='blue')
plt.title('Peak Departure Times')
plt.xlabel('Hour of Day')
plt.show()
```

```python
# 6. Price Distribution by Number of Stops
plt.figure(figsize=(10, 6))
sns.boxplot(x='stops', y='price', data=df, palette='Set2')
plt.title('Price Distribution by Number of Stops')
plt.show()
```

```python
# 7. Top 10 Most Expensive Flights
expensive_flights = df.nlargest(10, 'price')
print("Top 10 Most Expensive Flights:")
print(expensive_flights[['airline', 'from', 'to', 'price']])
```

```python
# 8. Average Flight Price by Airline
avg_price_airline = df.groupby('airline')['price'].mean().sort_values().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_airline.index, y=avg_price_airline.values, palette='pastel')
plt.title('Top 10 Airlines by Average Flight Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Class Distribution
plt.figure(figsize=(8, 6))
df['class'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Class Distribution')
plt.ylabel('')
plt.show()
```

```python
# 11. Cheapest Airlines
cheapest_airlines = df.groupby('airline')['price'].mean().sort_values().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=cheapest_airlines.index, y=cheapest_airlines.values, palette='viridis')
plt.title('Top 10 Cheapest Airlines by Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x='class', data=df, palette='Set3')
plt.title('Most Common Flight Class')
plt.show()
```

```python
# 13. Total Flights per Day
flight_per_day = df['flight date'].value_counts().sort_index()
plt.figure(figsize=(14, 6))
sns.lineplot(x=flight_per_day.index, y=flight_per_day.values, marker='o', color='teal')
plt.title('Total Flights per Day')
plt.xlabel('Date')
plt.ylabel('Number of Flights')
plt.xticks(rotation=45)
plt.show()
```

```python
# 14. Average Price Over Time
avg_price_time = df.groupby('flight date')['price'].mean()
plt.figure(figsize=(14, 6))
sns.lineplot(x=avg_price_time.index, y=avg_price_time.values, marker='o', color='red')
plt.title('Average Price Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# 15. Airline and Flight Class Pairing
sns.countplot(y='airline', hue='class', data=df, palette='husl')
plt.title('Airline and Flight Class Pairing')
plt.show()
```

```python
# Function to convert duration from 'xh ym' format to total minutes
def duration_to_minutes(duration):
    hours, minutes = 0, 0
    try:
        if 'h' in duration:
            parts = duration.split('h')
            hours = float(parts[0].strip())  # Allow for decimal hours
            if 'm' in parts[1]:
                minutes = int(parts[1].replace('m', '').strip())
        elif 'm' in duration:
            minutes = int(duration.replace('m', '').strip())
    except ValueError:
        print(f"Error processing duration: {duration}")  # Log any problematic durations
    return int(hours * 60 + minutes)  # Ensure return type is int

# Apply the function to convert duration to minutes
df['duration_minutes'] = df['duration'].apply(duration_to_minutes)

# Find the top 10 longest flights by duration in minutes
longest_flights = df.nlargest(10, 'duration_minutes')

# Display the top 10 longest flights
print("Top 10 Longest Flights by Route:")
print(longest_flights[['airline', 'from', 'to', 'duration']])
```

```python
# 18. Price Variability by Time of Departure
plt.figure(figsize=(14, 6))
sns.boxplot(x='dep_hour', y='price', data=df, palette='rainbow')
plt.title('Price Variability by Time of Departure')
plt.show()
```

```python
# 19. Most Frequent Destinations
top_destinations = df['to'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_destinations.index, y=top_destinations.values, palette='summer')
plt.title('Top 10 Most Frequent Destinations')
plt.xticks(rotation=45)
plt.show()
```
