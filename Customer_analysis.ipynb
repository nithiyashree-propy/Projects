```python

item_dim = pd.read_csv(r"C:\Users\nithi\Downloads\archive (10)\item_dim.csv", encoding='ISO-8859-1')
store_dim = pd.read_csv(r"C:\Users\nithi\Downloads\archive (10)\store_dim.csv", encoding='ISO-8859-1')
time_dim = pd.read_csv(r"C:\Users\nithi\Downloads\archive (10)\time_dim.csv", encoding='ISO-8859-1')
trans_dim = pd.read_csv(r"C:\Users\nithi\Downloads\archive (10)\Trans_dim.csv", encoding='ISO-8859-1')
customer_dim = pd.read_csv(r"C:\Users\nithi\Downloads\archive (10)\customer_dim.csv", encoding='ISO-8859-1')
```

```python
item_dim.to_string()
store_dim.to_string()
time_dim.to_string()
trans_dim.to_string()
customer_dim.to_string()
```

```python
print(item_dim.columns)
print(fact_table.columns)
print(store_dim.columns)
print(time_dim.columns)
print(trans_dim.columns)
print(customer_dim.columns)

```

```python
merged_data = pd.merge(fact_table, item_dim, on='item_key')
merged_data = pd.merge(merged_data, store_dim, on='store_key')
merged_data = pd.merge(merged_data, time_dim, on='time_key')
merged_data = pd.merge(merged_data, trans_dim, on='payment_key')
merged_data = pd.merge(merged_data, customer_dim, on='coustomer_key')
```

```python
print(merged_data.columns)
```

Total sales by store Division

```python
sales_by_division=merged_data.groupby('division')['total_price'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.bar(sales_by_division['division'],sales_by_division['total_price'])
plt.title('Total sales by store Division')
plt.xlabel('Store Division')
plt.ylabel('Total Sales')
plt.show()
```

Total Sales by month

```python
sales_by_month=merged_data.groupby('month')['total_price'].sum().reset_index()
plt.plot(sales_by_month['month'],sales_by_month['total_price'],marker='o')
plt.title('Total sales by month')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

Product price vs Total Sales

```python
merged_data.plot.scatter(x='unit_price_x', y='total_price', title='Unit Price vs Total Sales')
plt.show()
```

Customer Count by District


```python
customer_count_by_district = merged_data.groupby('district')['coustomer_key'].nunique().reset_index()
plt.bar(customer_count_by_district['district'], customer_count_by_district['coustomer_key'])
plt.title('Customer Count by District')
plt.xlabel('District')
plt.ylabel('Number of Customers')
plt.show()
```

Sales by payment type

```python
payment_sales = merged_data.groupby('trans_type')['total_price'].sum().reset_index()
plt.pie(payment_sales['total_price'], labels=payment_sales['trans_type'], autopct='%1.1f%%')
plt.title('Sales by Payment Type')
plt.show()
```

Total sales by product Category

```python
payment_sales = merged_data.groupby('trans_type')['total_price'].sum().reset_index()
plt.pie(payment_sales['total_price'], labels=payment_sales['trans_type'], autopct='%1.1f%%')
plt.title('Sales by Payment Type')
plt.show()
```

Total Sales by product category

```python
sales_by_category = merged_data.groupby('desc')['total_price'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.bar(sales_by_category['desc'], sales_by_category['total_price'])
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

Sales distribution across time of day

```python
sales_by_hour = merged_data.groupby('hour')['total_price'].sum().reset_index()
plt.bar(sales_by_hour['hour'], sales_by_hour['total_price'])
plt.title('Total Sales by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Total Sales')
plt.show()
```

top 10 products by sales volume

```python
top_products = merged_data.groupby('item_name')['total_price'].sum().nlargest(10).reset_index()
plt.bar(top_products['item_name'], top_products['total_price'])
plt.title('Top 10 Products by Sales Volume')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_sales_payment_type = merged_data.groupby('trans_type')['total_price'].mean().reset_index()
plt.bar(avg_sales_payment_type['trans_type'], avg_sales_payment_type['total_price'])
plt.title('Average Sales per Transaction by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Average Sales')
plt.show()
```

Customer retention analysis

```python
customer_purchases = merged_data.groupby('coustomer_key').size()
repeat_customers = (customer_purchases > 1).sum() / len(customer_purchases) * 100
new_customers = 100 - repeat_customers

plt.bar(['Repeat Customers', 'New Customers'], [repeat_customers, new_customers])
plt.title('Customer Retention Analysis')
plt.ylabel('Percentage')
plt.show()
```
