```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up plot style
sns.set(style="whitegrid")

# Load the data
# df = pd.read_csv('your_data.csv')  # Replace with your actual data source

# 1. Basic Descriptive Statistics
print(df.describe())
```

```python
# 2. Distribution Analysis for Income, Age, and Experience
for column in ['Income', 'Age', 'Experience']:
    sns.histplot(df[column], kde=True)
    plt.title(f'{column} Distribution')
    plt.show()
```

```python
# 3. Risk Flag Analysis
sns.countplot(x='Risk_Flag', data=df)
plt.title('Risk Flag Distribution')
plt.show()
```

```python
# 5. Age vs. Risk Flag
sns.violinplot(x='Risk_Flag', y='Age', data=df)
plt.title('Age by Risk Flag')
plt.show()
```

```python
# 6. Experience Analysis by Age Group
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-30', '30-40', '40-50', '50-60', '60+'])
sns.barplot(x='Age_Group', y='Experience', data=df, ci=None)
plt.title('Average Experience by Age Group')
plt.show()
```

```python
# 7. House Ownership and Risk
sns.countplot(x='House_Ownership', hue='Risk_Flag', data=df)
plt.title('House Ownership by Risk Flag')
plt.show()
```

```python
# 8. Marital Status Analysis
sns.countplot(x='Married/Single', hue='Risk_Flag', data=df)
plt.title('Marital Status by Risk Flag')
plt.show()
```

```python
# 10. Top Professions by Income
top_professions = df.groupby('Profession')['Income'].mean().sort_values(ascending=False).head(5)
sns.barplot(x=top_professions.index, y=top_professions.values)
plt.title('Top 5 Professions by Average Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# 11. City Analysis
top_cities = df['CITY'].value_counts().head(10)
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 10 Cities by Number of Individuals')
plt.xticks(rotation=45)
plt.show()
```

```python
# 12. State vs. Risk Flag
sns.countplot(x='STATE', hue='Risk_Flag', data=df, order=df['STATE'].value_counts().index[:10])
plt.title('State by Risk Flag (Top 10 States)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 13. Income vs. House Ownership
sns.boxplot(x='House_Ownership', y='Income', data=df)
plt.title('Income Distribution by House Ownership')
plt.show()
```

```python
# 14. Car Ownership and Risk
sns.countplot(x='Car_Ownership', hue='Risk_Flag', data=df)
plt.title('Car Ownership by Risk Flag')
plt.show()
```

```python
# 15. Income Analysis by Profession with Color Palette
prof_income = df.groupby('Profession')['Income'].mean().sort_values(ascending=False).head(10)
sns.barplot(x=prof_income.index, y=prof_income.values, palette='viridis')  # You can change 'viridis' to other palettes like 'coolwarm', 'Blues', etc.
plt.title('Top 10 Professions by Average Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# 16. Age Distribution by Car Ownership
sns.histplot(data=df, x='Age', hue='Car_Ownership', kde=True, multiple='stack')
plt.title('Age Distribution by Car Ownership')
plt.show()
```

```python
df['Experience_Group'] = pd.cut(df['Experience'], bins=5, labels=['Low', 'Medium-Low', 'Medium', 'Medium-High', 'High'])
sns.boxplot(x='Experience_Group', y='Income', hue='Risk_Flag', data=df, palette='coolwarm')
plt.title('Box Plot of Experience Groups vs. Income by Risk Flag')
plt.xticks(rotation=45)
plt.show()
```

```python
# 18. Current Job Years and Risk
sns.violinplot(x='Risk_Flag', y='CURRENT_JOB_YRS', data=df)
plt.title('Current Job Years by Risk Flag')
plt.show()
```

```python
# 19. Multiple Ownership Analysis
df['House_And_Car'] = (df['House_Ownership'] == 'Yes') & (df['Car_Ownership'] == 'Yes')
sns.countplot(x='House_And_Car', hue='Risk_Flag', data=df)
plt.title('House and Car Ownership by Risk Flag')
plt.show()
```

```python
# 20. Income vs. CURRENT_HOUSE_YRS
sns.scatterplot(x='CURRENT_HOUSE_YRS', y='Income', data=df)
plt.title('Income vs. Current House Years')
plt.show()
```

```python
# 1. Marital Status Distribution
df['Married/Single'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Marital Status Distribution')
plt.ylabel('')  # Hide the y-label for better visualization
plt.show()

# 2. House Ownership Distribution
df['House_Ownership'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightyellow'])
plt.title('House Ownership Distribution')
plt.ylabel('')
plt.show()

# 3. Car Ownership Distribution
df['Car_Ownership'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightpink', 'lightgray'])
plt.title('Car Ownership Distribution')
plt.ylabel('')
plt.show()

# 4. Risk Flag Distribution
df['Risk_Flag'].value_counts().plot(kind='pie', labels=['Low Risk', 'High Risk'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Risk Flag Distribution')
plt.ylabel('')
plt.show()
```
